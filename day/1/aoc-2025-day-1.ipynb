### Part 1

```python
def rotate(position, input):
    if input[0] == "R":
         magnitude = (int(input[1:]))
    else:
         magnitude = (0 - int(input[1:]))
    return (position + magnitude) % 100
```

```python
def readAndRun(position, file):
    pings = 0
    with open(file, 'r') as f:
        for line in f.readlines():
            position = rotate(position, line)
            if (position == 0):
                pings += 1
    return pings
```

```python
readAndRun(50, "input.txt")
```

### Part 2

```python
def clickLeft(position):
    position -= 1
    if position < 0:
        position = 99
    return position

def clickRight(position):
    position += 1
    if position > 99:
        position = 0
    return position
```

```python
def countPings(position, input):
    pings = 0
    clicks = (int(input[1:]))
    if input[0] == "R":
        right = True
    else:
        right = False
    for x in range(clicks):
        if right:
            position = clickRight(position)
        else:
            position = clickLeft(position)
        if position == 0:
            pings += 1
    return pings
```

```python
def readAndRun0x434C49434B(position, file):
    pings = 0
    with open(file, 'r') as f:
        for line in f.readlines():
            pings += countPings(position, line)
            position = rotate(position, line)
    return pings
```
