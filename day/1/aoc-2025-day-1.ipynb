```python
def rotate(position, input):
    if input[0] == "R":
         magnitude = (int(input[1:]))
    else:
         magnitude = (0 - int(input[1:]))
    return (position + magnitude) % 100
```

```python
def readAndRun(position, file):
    pings = 0
    with open(file, 'r') as f:
        for line in f.readlines():
            position = rotate(position, line)
            if (position == 0):
                pings += 1
    return pings
```

```python
readAndRun(50, "input.txt")
```
