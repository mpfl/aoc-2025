### Part 1

```python
def highestJoltage(inputBank):
    inputBankString = str(inputBank)
    if (len(str(inputBank))) == 2:
        return int(inputBank)
    else:
        highest = 0
        tens = 10 * int(inputBankString[0])
        for ones in inputBankString[1:]:
            test = tens + int(ones)
            if test > highest:
                highest = test
        return max(highest, int(highestJoltage(inputBankString[1:])))
```

```python
def readAndRun(file):
    joltages = []
    with open(file, 'r') as f:
        for line in f.readlines():
            joltages.append(highestJoltage(int(line)))
    return sum(joltages)
```

```python
readAndRun("input.txt")
```

```python
testBank1 = 987654321111111
testBank2 = 811111111111119
testBank3 = 234234234234278
testBank4 = 818181911112111

print( "Testing Bank 1: " )
expectedValue = 98
actualValue = highestJoltage(testBank1)
if (actualValue == expectedValue):
    print("Passed")
else:
    print("Failed: Expected " + str(expectedValue) + " but got " + str(actualValue))

print( "Testing Bank 2: " )
expectedValue = 89
actualValue = highestJoltage(testBank2)
if (actualValue == expectedValue):
    print("Passed")
else:
    print("Failed: Expected " + str(expectedValue) + " but got " + str(actualValue))
    
print( "Testing Bank 2: " )
expectedValue = 78
actualValue = highestJoltage(testBank3)
if (actualValue == expectedValue):
    print("Passed")
else:
    print("Failed: Expected " + str(expectedValue) + " but got " + str(actualValue))
    
print( "Testing Bank 2: " )
expectedValue = 92
actualValue = highestJoltage(testBank4)
if (actualValue == expectedValue):
    print("Passed")
else:
    print("Failed: Expected " + str(expectedValue) + " but got " + str(actualValue))
```
