### Part 1

```python
def isValidID(id):
    return not (str(id)[:int(len(str(id))/2)] == str(id)[int(len(str(id))/2):])
```

```python
def findInvalidIDs(inputRange):
    invalidIDs = []
    start = int(inputRange.split("-")[0])
    end = int(inputRange.split("-")[1])
    x = range(start, end + 1)
    for id in x:
        if not isValidID(id):
            invalidIDs.append(id)
    return invalidIDs
```

```python
def sumInvalidIDs(inputRanges):
    sumOfInvalids = 0
    invalidIDs = []
    inputRanges = inputRanges.split(",")
    for inputRange in inputRanges:
        sumOfInvalids += sum(findInvalidIDs(inputRange))
    return sumOfInvalids
```

```python
def readAndRun(file):
    with open(file, 'r') as f:
        for line in f.readlines():
            print(sumInvalidIDs(line))
```

```python
readAndRun("input.txt")
```

### Part 2

```python
def isValidID(id):
    returnValue = True
    factors = factorise(len(str(id)))
    for length in factors:
        splitID = list(map(''.join, zip(*[iter(str(id))]*length)))
        result = all(y == splitID[0] for y in splitID)
        if result:
            returnValue = False
    return returnValue
```

```python
def factorise(input):
    factors = []
    for x in range(1,int(input)):
        if (input % x) == 0:
            factors.append(x)
    return factors
```
